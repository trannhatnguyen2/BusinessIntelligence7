# REPORT

### 2021

```python
# filter sales in 2021 Q1
df_2021_Q1 = df_2021[(df_2021['transaction_date'] >= '2021-01-01') & (df_2021['transaction_date'] <= '2021-03-31')]
df_2021_Q1.info()
```

```python
#Pivot_table for 2021 Q1
df_pivot_2021 = df_2021_Q1.pivot_table(index='staff_id', values='sales', aggfunc='sum').sort_values(by='sales', ascending=False)
df_pivot_2021 = pd.merge(df_pivot_2021, df_empl[['first_name','last_name', 'staff_id']], on='staff_id', how='left')
df_pivot_2021.head()
```

COMMISSION OF EACH STAFF

```python
# Calculate commission for each staff
df_pivot_2021['commission'] = df_pivot_2021['sales'] * 0.05
#rearrage columns
df_pivot_2021 = df_pivot_2021[['staff_id', 'first_name', 'last_name', 'sales', 'commission']]
df_pivot_2021.head()
```

=> Britanni	Jorden is the staff having highets commission in Q1 2021

```python
# calculate percentage of sales for each staff
df_pivot_2021['sales_percentage'] = (round(df_pivot_2021['sales'] / sum(df_pivot_2021['sales'])*100,2)).astype(str) + '%'
df_pivot_2021
```

=> Britanni Jordan's contribution to the total sales was 12.02%.

```python
total_sales_2021_q1 = sum(df_pivot_2021['sales'])
total_sales_2021_q1
```

```python
len(df_pivot_2021['sales'])
```

```python
# Average sales per staff
avg_sales_2021 = total_sales_2021_q1 / len(df_pivot_2021['sales'])
avg_sales_2021
```
